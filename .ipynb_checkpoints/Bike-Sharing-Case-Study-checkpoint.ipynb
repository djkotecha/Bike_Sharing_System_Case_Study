```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
from sklearn.metrics import r2_score
```

```python
#load data file.

bike = pd.DataFrame(pd.read_csv("day.csv"))
bike.head()
```

```python
# Duplicate row check
bike.drop_duplicates(subset=None, inplace=True)
bike.shape

# Observations:

# There is no change in number of rows, so no duplicates found.
```

```python
# Data Cleaning

temp_bike = bike.iloc[:, 1:16]

temp_bike.head()
```

```python
for col in temp_bike:
    print(temp_bike[col].value_counts(ascending=False), '\n')
    
# Observation:
# Data seems to be good. 
```

```python
# Remove unwanted columns

bike.head()

# instant - is just index columns, we don't need it for analysis
# dteday - is date column but we have month and year columns too. We are not analysing on particular date, so we can live with 
# month and year columns. So let's remove dteday also.
# Casual and Registered - are categories, but we want to analyse the total count and we have "cnt" column for that. So we can 
# drop these two columsn also.
```

```python
# Let's drop unwanted columns

bike_new = bike.drop(['instant','dteday', 'casual', 'registered'],axis=1)
#bike_new.drop(['instant', 'dteday', 'casual', 'registered'], axis=1, inplace=True)
bike_new.head()
```

## Create Dummy Variables

```python
# Convert categorical variables to category datatype for the once having more than 2 categories.

bike_new['season']=bike_new['season'].astype('category')
bike_new['weathersit']=bike_new['weathersit'].astype('category')
bike_new['mnth']=bike_new['mnth'].astype('category')
bike_new['weekday']=bike_new['weekday'].astype('category')

bike_new.info()
```

```python
bike_new = pd.get_dummies(bike_new, drop_first=True)
bike_new.info()

# Observations:
# This creates dummy variables for category variables and drops first dummy created for each category as well as the original variable.
```

## Split the data between train and test data

```python
np.random.seed(0)
df_train, df_test = train_test_split(bike_new, train_size = 0.70, test_size = 0.30, random_state = 100)
```

## Re-Scaling The Features

```python
# We will apply scaling to numeric columns. So list them separately.
lst_numeric = ['temp', 'atemp', 'hum', 'windspeed','cnt']

# Create Scaler Object
scaler = MinMaxScaler()

df_train[lst_numeric] = scaler.fit_transform(df_train[lst_numeric])
```

```python
df_train.head()
```

## Build A Linear Model

```python
# First divide the data frame in x and y sets of data

y_train = df_train.pop('cnt')
X_train = df_train
```

## RFE

```python
# We will use Recursive Feature Elimination method
# We have 30 columns (variables). Let's start with 15 variables out of 30.
# Create an object of LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, step = 15)
rfe = rfe.fit(X_train, y_train)
```

```python
# Let's see the which 15 columns participated

col = X_train.columns[rfe.support_]
col
```

```python
# Let's create a data frame with RFE selected variables

X_train_rfe = X_train[col]

# we will use this for VIF check for manual model creation using Stats Model
```

## Build Linear Model using 'Stats Model'

#### Model 1

```python
# VIF check of feature variables

vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif.head(15)
```

```python
# stats model doesn't have the constant, so we will add it.

X_train_sm1 = sm.add_constant(X_train_rfe)

# Create first fitted model
lr1 = sm.OLS(y_train, X_train_sm1).fit()

# Check obtained parameters
lr1.params
```

```python
print(lr1.summary())

# Observations:
# P-value for most of the variables is 0. so the result with most variables is significant.
# p-value for mnth_4 and mnth_5 are higher, i.e. > 5% so the result is not significant for these 2 variables in this model.
# R-squared = 0.841, that means about 84% of variance in cnt is explained by the participating variables, which is good.
```

#### Model 2

```python
# removing mnth_4 based on high p-value

X_train_new = X_train_rfe.drop(["mnth_4"], axis = 1)
```

```python
# VIF check after removing "mnth_4"
vif = pd.DataFrame()
vif['Features'] = X_train_new.columns
vif['VIF'] = [variance_inflation_factor(X_train_new.values, i) for i in range(X_train_new.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif.head(15)
```

```python
# Add a constant
X_train_sm2 = sm.add_constant(X_train_new)

# Create a first fitted model
lr2 = sm.OLS(y_train, X_train_sm2).fit()
```

```python
# Check obtained parameters
lr2.params
```

```python
# Print a summary of the linear regression model obtained
print(lr2.summary())

# Observations:
# P-value for most of the variables is 0. so the result with most variables is significant.
# p-value for mnth_10 is < 5% So the result for those variables is also significant. But we can consider
#     removing one of them in next model.
# p-value for mnth_5 has increased compared to previous model.
# R-squared = 0.840, that means about 84% of variance in cnt is explained by the participating variables, which is good.
```

## Final Model Analysis

```python
# As per Hypothesis testing:
# Our null hypotheis is H0:B1=B2=...=Bn=0 and
#                       H1: at least one Bi!=0

# As per the final model summary, we can say that none of the variable's coefficient are zero.
# That mean "We Reject The Null Hypotheis".
```

#### F Statistics

```python
# F-Statistics is used for testing the overall significance of the Model: Higher the F-Statistics, more significant the Model is.

# F-statistic: 270.8
# Prob (F-statistic): 6.57e-186
# The F-Statistics value of 270.8 (which is greater than 1) and the p-value of '~0.0000' states that the overall model is significant
```

## Assumptions

#### 1. Error terms are normally distributed with mean zero (not X, Y)

```python
# Residual Analysis Of Training Data

y_train_pred = lr6.predict(X_train_sm6)
```

```python
res = y_train-y_train_pred
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((res), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)
plt.xlabel('Errors', fontsize = 18)

# Observation
# From the below histogram, we could see that the Residuals are normally distributed. 
# Hence our assumption for Linear Regression is valid.
```

####  2. There is a linear relationship between X and Y

#### 3. There is No Multicollinearity between the predictor variables

```python
# VIF results of last model
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train_new.columns
vif['VIF'] = [variance_inflation_factor(X_train_new.values, i) for i in range(X_train_new.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif

# Observations:
# From the VIF calculation we could find that there is no multicollinearity existing between the predictor variables,
# as all the values are within permissible range of 5 or below 5
```

## Evaluate The Model

```python
# Visualise the y_test and y_pred on scatter chart

fig = plt.figure()
plt.scatter(y_test, y_pred, alpha=.5)
fig.suptitle('y_test vs y_pred', fontsize = 20)              
plt.xlabel('y_test', fontsize = 18)                         
plt.ylabel('y_pred', fontsize = 16) 
plt.show()
```

## R^2 Value for Test

```python
r2 = r2_score(y_test, y_pred)
r2
```

## Adjusted R^2 Value for Test

```python
# We have r2 above

r2 = 0.7909416676310566

# Shape of test set X_test
X_test.shape
```

```python
# n is number of rows in X

n = X_test.shape[0]

# Number of features (predictors, p) is the shape along axis 1
p = X_test.shape[1]

# We find the Adjusted R-squared using the formula

adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
adjusted_r2
```
